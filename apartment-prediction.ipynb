```python
import pandas as pd
import numpy as np

apartment_data = pd.read_csv("apartment_data.csv")
apartment_data
```

```python
average_apartment_price = round(apartment_data.price.mean())
print(f"Average apartment price: {average_apartment_price}")
```

```python
# make location show only one location, like Westlands
apartment_data.loc[:, "location"] = (
    apartment_data.loc[:, ("location")].str.split(",").str[-1].str.strip()
)
apartment_data
```

```python
# remove empty rows
apartment_data = apartment_data.dropna(axis=0)
apartment_data
```

```python
# split external features into new fields
external_features = apartment_data["external_features"].str.split(",")

for feature in external_features.explode().unique():
    apartment_data.loc[:, (f"external_has_{feature.strip()}")] = (
        apartment_data.loc[:, ("external_features")]
        .apply(lambda x: 1 if feature.strip() in x else 0)
    )

apartment_data.head()
```

```python
# split internal features into new fields
internal_features = apartment_data["internal_features"].str.split(",")

for feature in internal_features.explode().unique():
    apartment_data.loc[:, (f"internal_has_{feature.strip()}")] = apartment_data.loc[
        :, ("internal_features")
    ].apply(lambda x: 1 if feature.strip() in x else 0)

apartment_data.head()
```

```python
# split internal features into new fields
nearby_features = apartment_data["nearby"].str.split(",")

for feature in nearby_features.explode().unique():
    apartment_data.loc[:, (f"nearby_{feature.strip()}")] = (
        apartment_data.loc[:, ("nearby")]
        .apply(lambda x: 1 if feature.strip() in x else 0)
        .copy()
    )

apartment_data.head()
```

```python
# unique locations
unique_locations = apartment_data["location"].unique()
len(unique_locations)
```

```python
# convert location values into numerical values
location_labels, loaction_uniques = pd.factorize(apartment_data["location"])
apartment_data.loc[:, "location_encoded"] = location_labels
apartment_data.head()
```

```python
# the target prediction is price
y = apartment_data["price"]

# the features will include bedrooms, bathrooms and location
external_features = [x for x in apartment_data.columns if "external_has_" in x]
external_features
apartment_features = ["bedrooms", "bathrooms", "location_encoded", [x for x in external_features]]
apartment_features
# X = apartment_data[apartment_features]

# X.tail(100)
```

```python
# split external features into new fields
external_features = apartment_data["external_features"].str.split(",")

for feature in external_features.explode().unique():
    apartment_data.loc[:, (f"external_has_{feature.strip()}")] = (
        apartment_data.loc[:, ("external_features")]
        .apply(lambda x: 1 if feature.strip() in x else 0)
    )

apartment_data.head()
```

```python
# put the external, internal and nearby features into a single list with all the features being used to train the model
import itertools

external_features = [x for x in apartment_data.columns if "external_has_" in x]
internal_features = [x for x in apartment_data.columns if "internal_has_" in x]
nearby_features = [x for x in apartment_data.columns if "nearby_" in x]

apartment_features = [
    "bedrooms",
    "bathrooms",
    "location_encoded",
]
apartment_features.extend(external_features)
apartment_features.extend(internal_features)
apartment_features.extend(nearby_features)
apartment_features
```

```python
# set the features we're going to use
X = apartment_data[apartment_features]
```

```python
from sklearn.tree import DecisionTreeRegressor

apartment_model = DecisionTreeRegressor()
apartment_model.fit(X, y)
```

```python
print("First in-sample predictions:", apartment_model.predict(X.head(10)))
print("Actual target values for those homes:", y.head(10))
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
apartment_model = DecisionTreeRegressor(random_state=1)
apartment_model.fit(train_X, train_y)
```

```python
value_predictions = apartment_model.predict(val_X)
```

```python
print(value_predictions[0:10])
print(list(y.head(10)))
```

```python
from sklearn.metrics import mean_absolute_error

value_mae = mean_absolute_error(val_y, value_predictions)
print(f"Mean Absolute Error: {value_mae}")
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae
```

```python
leaf_size_candidates = [5, 25, 50, 100, 250, 500, 750, 1000]

mae_scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in leaf_size_candidates}
mae_scores
```

```python
best_tree_size = min(mae_scores, key=mae_scores.get)
print(f"Best tree size: {best_tree_size} \t Mean Absolute Error: {mae_scores[best_tree_size]}")
```

```python
final_decision_tree_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)
final_decision_tree_model.fit(X, y)
```

```python
print(final_decision_tree_model.predict(X.head(10)))
print(list(y[0:10]))
```

Now we do this using `RandomForestRegressor` instead of `DecisionTreeRegressor`.

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
apartment_predictions = forest_model.predict(val_X)
print(f"Mean absolute error: {mean_absolute_error(val_y, apartment_predictions)}")
```
